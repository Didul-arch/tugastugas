### 1. Dilasi
Definisi : Teknik yang digunakan untuk menebalkan tepi citra \
Output : Tepi citra lebih tebal

![Dilasi image](https://miro.medium.com/v2/resize:fit:720/format:webp/1*PGoNvgGjpPKE8PLFkUZm2w.png "Gambar")

Note : Jika ukuran **SE semakin besar, citra semakin tebal**.


Pixel (output) = 1 {if HIT} \
Pixel (output) = 0 {otherwise}

![Dilasi image](https://miro.medium.com/v2/resize:fit:720/format:webp/1*GcufLGivTUHf-wDic2GeVw.png "MarineGEO logo")

![Gambar](https://miro.medium.com/v2/resize:fit:720/format:webp/1*fh4Ncpdl4ZI1w-icV9AyxQ.png)

### 2. Erosi
Definisi : Teknik yang digunakan untuk menipiskan tepi citra \
Output : Tepi citra lebih tipis

![gambar](https://miro.medium.com/v2/resize:fit:720/format:webp/1*uYB8B-g5ZkuRCUlzi0OgQQ.png)

Jika ukuran **SE semakin besar, tepi citra semakin tipis**.


Pixel (output) = 1 {if FIT}\
Pixel (output) = 0 {otherwise}

![gambar](https://miro.medium.com/v2/resize:fit:720/format:webp/1*6i7JdXdnKVjkNA1pH3xZ0g.png)

![gambar](https://miro.medium.com/v2/resize:fit:720/format:webp/1*KsKoNMXRzeJCVmiOp7u0TA.png)

```python
cv2_imshow(img)

kernel = np.ones((6,6), np.uint8)
erosi = cv2.erode(img, kernel, iterations = 2)
cv2_imshow(erosi)
```

### 3. Opening
Definisi : Erosi --> Dilation \
Opening berguna untuk **menghaluskan citra,
menghilangkan tonjolan yang tipis, memisahkan objek dan menghilangkan noise**.

![image](https://miro.medium.com/v2/resize:fit:720/format:webp/1*YDeccRxRQsvSCx4eSmKqFw.png)

```python
# Program closing
img = cv2.imread('/content/drive/MyDrive/PCD-Prak/melon_noise.png')
cv2_imshow(img)
```

```python
kernel = np.ones((15,15), np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
cv2_imshow(opening)
```

### 4. Closing
Definisi : Dilation --> Erosi \
Closing berguna untuk menghaluskan citra serta menghilangkan lubang kecil
![image](https://miro.medium.com/v2/resize:fit:720/format:webp/1*YDeccRxRQsvSCx4eSmKqFw.png)

```python
# Program closing
img = cv2.imread('/content/drive/MyDrive/PCD-Prak/closing.png', 0)
cv2_imshow(img)
```

```python
kernel = np.ones((15,15), np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
cv2_imshow(closing)
```
